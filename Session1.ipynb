```python
import seaborn as sns
```

```python
data = sns.load_dataset("tips")    # Load: the data from github
data
```

```python
sns.lineplot(x="total_bill", y="tip", data = data)
```

```python
sns.lineplot(x="total_bill", y="tip", data = data, hue="sex")
```

```python
sns.lineplot(x="total_bill", y="tip", data = data, hue="sex", style = "sex")
```

```python
sns.barplot(x="day", y="total_bill", data=data, hue="day")
```

```python
sns.scatterplot(x="total_bill", y="tip", data=data, hue="sex", style="sex")
```
