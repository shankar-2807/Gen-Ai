```python
plt.plot(iris_setosa["Petal_Length"], np.zeros_like(iris_setosa["Petal_Length"]), "o", label = "Setosa")
plt.plot(iris_virginica["Petal_Length"], np.zeros_like(iris_virginica["Petal_Length"]), "o",label="Virginica")
plt.plot(iris_versicolor["Petal_Length"], np.zeros_like(iris_versicolor["Petal_Length"]), "o", label="Versicolor")      # Versicolor
```

```python
plt.plot(iris_setosa["Petal_Length"], np.zeros_like(iris_setosa["Petal_Length"]), "o", label = "Setosa")
plt.plot(iris_virginica["Petal_Length"], np.zeros_like(iris_virginica["Petal_Length"]), "o",label="Virginica")
plt.plot(iris_versicolor["Petal_Length"], np.zeros_like(iris_versicolor["Petal_Length"]), "o", label="Versicolor")    
plt.xlabel("Petal Length")
plt.title("Petal length of Iris species")
plt.legend()
plt.show()
```

```python
sns.histplot(data=iris, x = "Petal_Length", hue = "Species", element="step", stat="density", common_norm=False)
```

```python
sns.histplot(data=iris, x = "Petal_Length", hue = "Species", element="step", stat="density", common_norm=False)
plt.title("Histogram of petal length by iris species")
plt.xlabel("Patel Length")
plt.ylabel("Density")
plt.show()
```

```python
sns.histplot(data=iris, x="Petal_Length", hue="Species",common_norm=False)
sns.kdeplot(data=iris, x="Petal_Length", hue="Species", common_norm=False, fill=False)
```
